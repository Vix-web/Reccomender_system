# Fetching the video data from the given API

```python
def fetch_all_videos(api_url):
    videos = []
    page = 1

    while True:
        response = requests.get(f"{api_url}?page={page}")
        if response.status_code == 200:
            data = response.json()
            if 'posts' in data and data['posts']:
                videos.extend(data['posts'])
                page += 1
            else:
                break
        else:
            print(f"Failed to fetch data: {response.status_code}")
            break
    return videos
video_data = fetch_all_videos("https://api.socialverseapp.com/feed")
```

##TF-IDF vectorization and using Truncated SVD

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import pairwise_distances
```

```python
tfidf = TfidfVectorizer(token_pattern="[a-z']+", max_df=0.9, lowercase=True)
```

Created an instance TfidfVectorizer that  converts the contents into a matrix of TF-IDF features. The token_pattern restricts tokens to lowercase letters and apostrophes, and max_df=0.9 filters out tokens that appear in more than 90% of documents. Also converts uppercase letters to lowercase to standardize the vocabulary.

```python
tfidf_matrix = tfidf.fit_transform(contents)
tfidf_matrix.shape   #Checking the matrix shape
```

```python
svd = TruncatedSVD(n_components=50, algorithm='arpack') #initializing truncatedSVD
```

Initialized TruncatedSVD to reduce the high dimensional TF-IDF matrix to 50 components(lower dimension).

```python
svd_mat = svd.fit_transform(tfidf.fit_transform(contents)) # svd_mat represents the content features in a lower-dimensional space.
```

```python
svd_mat_df = pd.DataFrame(svd_mat, index=video_data_new.id)
svd_mat_df.head()
```

Converted svd_mat into a DataFrame with video IDs as the index.

```python
def _combine_videos(list_of_video_ids):
    return svd_mat_df.loc[eval(list_of_video_ids),:].values.mean(axis=0)
```

This function that takes a list of video IDs (from a user‚Äôs watch history), retrieves their corresponding SVD-reduced vectors from svd_mat_df, and then computes the mean vector. This represents the user‚Äôs overall watch history as a single vector.

```python
lang_map = {'English':3, 'Spanish':2, 'French':1}
loc_map = {'Spain':3, 'USA':2, 'France':1}
```

```python
user_data_2 = user_data.copy() #creating a copy to keep the original data
```

```python
user_data_2.head()
```

```python
user_data_3 = np.hstack([user_data_2[['language','location']].values, np.array(user_watch_history_vector)])
```

```python
user_data_3
```

In a matrix user_data_3, each row represents a user with both content-based (watch history) and context-based (language, location) features.

```python
def _get_recommendations(user_name, top_k_users=5, top_k_vid=5):
    user_row = user_data[user_data['name'] == user_name]
    if user_row.empty:
        return f"User '{user_name}' not found in user_data."

    watched_vid_list = set(eval(user_row['watch_history'].values[0]))

    # Computing the distance of this user to all other users
    dist = pairwise_distances(X=user_data_3,Y=user_data_3[user_row.index[0], :].reshape(1, -1),metric='cosine')

    # Finding the top K most similar users
    top_k_similar_users = np.argsort(dist.flatten())[1:top_k_users + 1]

    # Collecting the watch histories of the top K similar users
    vid_ids_similar_users = []
    for user in top_k_similar_users:
        vid_ids_similar_users += eval(user_data.iloc[user]['watch_history'])

    # Removing videos that the target user has already seen
    unique_vid_ids_similar_users = set(vid_ids_similar_users)
    unseen_vid_ids_by_query_user = list(unique_vid_ids_similar_users - watched_vid_list)

    # Fetching the top K videos based on the inspiration score
    df = video_data[video_data['id'].isin(unseen_vid_ids_by_query_user)][['id', 'inspiration_score']]
    df = df.sort_values('inspiration_score', ascending=False).head(n=top_k_vid)

    return df
```

```python
recommendations = _get_recommendations(user_name='Robbin Stanback', top_k_users=5, top_k_vid=5)
recommendations
```

The function above recommends videos to a user by finding other users with similar tastes, gathering their watch histories, filtering out videos the user has already seen, and then selecting the best remaining videos based on their ratings. The recommendations are personalized to each user based on their preferences and the preferences of similar users.

The 'inspiration_score'(average rating) is used here to sort the top k videos because it is a metric that likely reflects the overall quality or appeal of a video. By using it to rank and filter recommended videos, the system aims to present users with content that they are more likely to find engaging and enjoyable. This enhances the user experience by ensuring that the recommended videos are not just relevant, but also of high quality.

```python
def generate_all_recommendations(top_k_users=5, top_k_vid=5):
    recommendations = {}

    for idx, row in user_data.iterrows():
        user_name = row['name']
        recommended_videos = _get_recommendations(user_name, top_k_users=top_k_users, top_k_vid=top_k_vid)
        recommendations[user_name] = recommended_videos['id'].tolist()
    return recommendations
```

The above function is designed to generate video recommendations for every user in the 'user_data' DataFrame. It creates a dictionary where each key is a user's name, and the value is a list of video IDs recommended for that user.
